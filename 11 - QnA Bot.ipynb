# IA conversationnelle

Pensez à la fréquence à laquelle vous communiquez avec d’autres personnes par le biais de la messagerie instantanée, des réseaux sociaux, des e-mails ou d’autres technologies en ligne. Pour nombre d’entre nous, c’est le moyen de communication par excellence. Lorsque vous avez une question au travail, vous pouvez contacter un collègue par le biais de la messagerie instantanée, accessibles sur vos appareils mobiles, afin d’être toujours en contact.

![Un humain et un robot en pleine conversation](./images/conversational_ai.jpg)

Les bots sont des agents d’intelligence artificielle qui communiquent par ce type de canaux, permettant un engagement naturel et conversationnel avec les services logiciels.

## Créer une base de connaissances QnA Maker

Pour les scénarios service clientèle, il est courant de créer un bot capable d’interpréter les questions fréquemment posées et d’y répondre par le biais d’une fenêtre de conversation sur un site web, d’un email ou d’une interface vocale. L’interface du bot repose sur une base de connaissances comportant des questions et des réponses appropriées que le bot peut consulter pour trouver les réponses adéquates.

Le service QnA Maker est un service cognitif d’Azure qui permet de créer rapidement une base de connaissances, soit en saisissant des paires de questions et de réponses, soit à partir d’un document ou d’une page web existants. Il peut ensuite utiliser certaines fonctionnalités intégrées de traitement du langage naturel pour interpréter les questions et trouver des réponses appropriées.

1. Ouvrez un autre onglet de navigateur et accédez au portail QnA Maker à l’adresse https://qnamaker.ai. Connectez-vous avec le compte Microsoft associé à votre abonnement Azure.
2. Dans le portail QnA Maker, sélectionnez **Créer une base de connaissances**.
3. Si vous n’avez pas encore créé de ressource de service QnA, sélectionnez **Créer un service QnA**. Le portail Azure s’ouvre sous un autre onglet pour vous permettre de créer un service QnA Maker dans votre abonnement. Utilisez les paramètres suivants :
   - **Géré (aperçu)** : Non sélectionné. 
   - **Abonnement** : *Votre abonnement Azure*
   - **Groupe de ressources** : *Sélectionnez un groupe de ressources existant ou créez-en un*
   - **Nom** : *Nom unique de votre ressource QnA*
   - **Niveau tarifaire** : F0
   - **Niveau tarifaire de la Recherche Azure** : F
   - **Emplacement de la Recherche Azure** : *N’importe quel emplacement disponible*
   - **Nom de l’application** : _Identique à la valeur  **Nom** (« .azurewebsites.net » sera ajouté automatiquement)_
   - **Emplacement du site web** : _Identique à la valeur **Emplacement de la Recherche Azure**_
   - **App Insights** : Désactiver

> **Remarque** : Si vous avez déjà approvisionné un **QnA Maker** de niveau Gratuit ou des ressources **Recherche Azure**, il se peut que votre quota ne vous permette pas d’en créer un autre. Dans ce cas, sélectionnez un niveau différent de **F0** / **F**.

4. Attendez que le déploiement du service QnA et des ressources associées se termine dans le portail Azure.
5. Revenez à l’onglet du portail QnA Maker. Dans la section **Étape 2**, cliquez sur **Actualiser** pour actualiser la liste des ressources de service QnA disponibles.
6. Connectez votre service QnA à votre base de connaissances en sélectionnant les options suivantes :
  - **ID Microsoft Azure Active Directory** : *L’ID répertoire Azure de votre abonnement*
  - **Nom de l’abonnement Azure** : *Votre abonnement Azure*
  - **Azure QnA Service** : *La ressource service QnA que vous avez créée dans la section précédente*
  - **Langue** : Français

### (!) Vérification
Si un message indiquant que le rôle n’a pas la permission d’effectuer l’action s’affiche, actualisez la page du navigateur pour le portail QnA Maker.

7. Dans la section **Étape 3**, saisissez le nom **Base de connaissances Margie’s Travel**.
8. Dans la section **Étape 4**, dans la zone **URL**, tapez <a href=’‘>*https://github.com/MicrosoftDocs/ai-fundamentals/raw/master/data/qna_bot/margies_faq.docx*</a>, puis cliquez sur **+ Ajouter URL**. Ensuite, sous **Chit-chat**, sélectionnez **Professionnel**.
9. Dans la section **Étape 5**, cliquez sur **Créer votre base de connaissances**.
10. Patientez pendant la création de votre base de connaissances, qui peut prendre une minute ou deux. Passez ensuite en revue les questions et les réponses qui ont été importées du document FAQ et les réponses prédéfinies du chat professionnel.

## Éditer la base de connaissances

Votre base de connaissances est basée sur les détails du document FAQ et sur certaines réponses prédéfinies. Vous pouvez ajouter des paires de questions et réponses personnalisées pour les compléter.

1. Cliquer sur **+ Ajouter une paire QnA**.
2. Dans la zone **Question**, tapez `Hello`. Cliquez ensuite sur **+ Ajouter une formulation alternative** et saisissez `Hi`.
3. Dans la zone **Réponse**, tapez `Hello`.

## Former et tester la base de connaissances

Maintenant que vous disposez d’une base de connaissances, vous pouvez la tester dans le portail QnA Maker.

1. Dans le coin supérieur droit de la page, cliquez sur **Enregistrer et former** pour former votre base de connaissances. Vous devrez peut-être agrandir la fenêtre pour voir le bouton.
2. Une fois la formation terminée, cliquez sur **&larr; Test** pour ouvrir le volet de test.
3. Dans la partie inférieure du volet de test, tapez le message *Salut*. La réponse **Bonjour** doit être renvoyée.
4. Dans la partie inférieure du volet de test, saisissez le message *Je veux réserver un vol*. Une réponse appropriée du FAQ doit être renvoyée.
5. Lorsque vous avez terminé de tester la base de connaissances, cliquez sur **&rarr; Test** pour fermer le volet de test.

## Créer un bot pour la base de connaissances

La base de connaissances fournit un service de bout en bout que les applications clientes peuvent utiliser pour répondre aux questions par le biais d’une sorte d’interface utilisateur. En règle générale, ces applications clientes sont des bots. Pour mettre la base de connaissances à la disposition d’un bot, vous devez la publier en tant que service accessible via HTTP. Vous pouvez ensuite utiliser le service Azure Bot pour créer et héberger un bot qui utilise la base de connaissances pour répondre aux questions des utilisateurs.

1. Dans la partie supérieure de la page QnA Make, cliquez sur **Publier**. Dans la **Base de connaissances Margies Travel**, cliquez ensuite sur **Publier**.
2. Une fois le service déployé, cliquez sur **Créer un bot**. Le portail Azure s’ouvre alors sous un nouvel onglet de navigateur et vous permet de créer un bot Web App dans votre abonnement Azure.
3. Dans le portail Azure, créez un bot Web App avec les paramètres suivants (la plupart seront automatiquement pré-remplis) :
  - **Identification du bot** : *Nom unique de votre bot*
  - **Abonnement** : *Votre abonnement Azure*
  - **Groupe de ressources** : *Groupe de ressources contenant votre ressource QnA Maker*
  - **Emplacement** : *Emplacement identique à celui de votre service QnA Maker*.
  - **Niveau tarifaire** : F0
  - **Nom de l’application** : *Identique à la valeur **Identification du bot**, où *.azurewebsites.net* est ajouté automatiquement
  - **Langage du SDK** : *Choisissez soit C# soit Node.js*
  - **Clé d’authentification QnA** : *Automatiquement défini sur la clé d’authentification de votre base de connaissances QnA*
  - **Plan/emplacement App Service** : *Automatiquement défini sur le plan et l’emplacement appropriés*
  - **Application Insights** : Désactivé
  - **ID d’application et mot de passe Microsoft** : Création automatique de l’ID d’application et du mot de passe.
4. Patientez pendant la création de votre bot (une icône de notification, représentant une cloche, s’affiche en haut à droite). Ensuite, dans la notification indiquant que le déploiement est terminé, cliquez sur **Accéder à la ressource** (ou, sur la page d’accueil, cliquez sur **Groupes de ressources*, ouvrez le groupe de ressources où vous avez créé le bot de l’application web et cliquez dessus).
5. Dans le panneau de votre bot, affichez la page **Tester dans une discussion web** et attendez que le bot affiche le message **Bonjour et bienvenue** (l’initialisation peut prendre quelques secondes).
6. Utilisez l’interface de discussion test pour vous assurer que votre bot répond aux questions de votre base de connaissances comme prévu. Par exemple, essayez d’envoyer *Je dois annuler mon hôtel*.
